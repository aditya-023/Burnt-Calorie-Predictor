```python
import numpy as np
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import xgboost
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
print("numpy", np.__version__)
print("pandas", pd.__version__)
print("matplotlib", mp.__version__)
print("seaborn", sns.__version__)
print("sklearn", sklearn.__version__)
print("xgboost", xgboost.__version__)
```

Data Collection and Processing

```python
#Loading the data from a csv file to a Pandas data frame
calories=pd.read_csv('/content/calories.csv')
```

```python
#Printing the first 5 rows of the data frame
calories.head()
```

```python
exercise_data=pd.read_csv('/content/exercise.csv')
```

```python
exercise_data.head()
```

Combining the 2 data frames

```python
calories_data=pd.concat([exercise_data,calories['Calories']],axis=1)
```

```python
calories_data.head()
```

```python
#Checking the number of rows and columns
calories_data.shape
```

```python
#Getting important info about the data
calories_data.info()
```

```python
#Checking for missing values
calories_data.isnull().sum()
```

Data Analysis

```python
#Some statisitical measures about the data
calories_data.describe()
```

Data visualisation

Finding the correlation in the data

 1.Positive Correlation

 2.Negative Correlation

```python
calories_data_numeric = pd.get_dummies(calories_data, drop_first=True)
correlation = calories_data_numeric.corr()
```

```python
# Constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar= True, square= True,fmt= '.1f',annot= True, annot_kws= {'size' :8},cmap= 'Blues')
```

Converting the text data into numerical values

```python
calories_data.replace({"Gender": {'male': 0, 'female': 1}}, inplace=True)
```

```python
calories_data.head()
```

Separating features and target

```python
X = calories_data.drop(columns=['User_ID','Calories'],axis=1)
Y = calories_data['Calories']
```

```python
print(X)
```

```python
print(Y)
```

Performing Traint-Test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

XGBoost Regressor

```python
#Loading the model
model = XGBRegressor()
```

```python
#Training the model with X_train
model.fit(X_train,Y_train)
```

Evaluation of the model

Prediction on test data


```python
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

Mean Absolute Error

```python
mae=metrics.mean_absolute_error(Y_test,test_data_prediction)
```

```python
print("Mean Absolute Error =", mae)
```
